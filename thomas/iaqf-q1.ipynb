## Data

```python
info = pd.read_csv('constituents.csv')
```

```python
sectors = info['GICS Sector'].unique()
```

```python
df = pd.read_csv("SP500.csv")
df = df.set_index('Date')
benchmark = yf.download(tickers = ['SPY'],start = date(2022,1,1),end = date(2025,1,1))['Close']
```

```python
df.index = pd.to_datetime(df.index).date
```

Clean

```python
df = df.drop(columns = ['BF.B','LEN', 'BRK.B','CEG', 'GEHC', 'GEV', 'KVUE', 'SOLV', 'SW', 'VLTO'])
df = np.log(df).diff().dropna()

benchmark = np.log(benchmark).diff().dropna()
```

#### Rolling Betas

```python
rolling_window = 120

# Step 4: Calculate rolling covariance and variance for each stock
rolling_cov = df.rolling(window=rolling_window).cov(benchmark['SPY'])

rolling_var = benchmark['SPY'].rolling(window=rolling_window).var()

# Step 5: Calculate rolling beta for each stock
rolling_beta = rolling_cov.div(rolling_var, axis=0)
```

```python
fig = go.Figure()

for sector in sectors:
    tickers = info.loc[info['GICS Sector'] == sector,'Symbol'].unique()
    
    sector_mean = rolling_beta.filter(items=tickers, axis=1).mean(axis = 1).dropna()
    
    fig.add_trace(
        go.Scatter(
            x = sector_mean.index,
            y = sector_mean.values,
            name = sector
        ),
        
    )

fig.update_layout(title = 'Beta by Sector (120d Rolling)')
fig.show()
```

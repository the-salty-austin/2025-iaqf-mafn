```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
from datetime import date
```

```python
mc = yf.download(tickers= ['SPY','^SP600'],start = date(2010,1,1))['Close']
# ew = yf.download(tickers = ['SPXEW','^R2ESC'],start = date(2020,1,1),period ='5y')['Close']
spy = yf.download(tickers = ['SPY'],start = date(2010,1,1))['Close']
```

```python
mc = mc.pct_change().dropna()
spy = spy.pct_change().dropna()
# ew = ew.pct_change().dropna()
# spy = (1+ spy['SPY']).cumprod()

mc = mc['^SP600'] - mc['SPY']
# ew = (1 + ew['^R2ESC'] - ew['SPXEW']).cumprod()
```

```python
rolling_corr = mc.rolling(252).corr(spy)
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = rolling_corr.index,
        y = rolling_corr['SPY'],
        name = 'Corr (252)'
    )
)

fig.update_layout(title = 'Rollign Corr (Small - Large) vs SPY')
fig.show()
```

```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
from datetime import date
```

```python
mc = yf.download(tickers= ['SPY','^SP600'],start = date(2020,1,1))['Close']
ew = yf.download(tickers = ['SPXEW','^R2ESC'],start = date(2020,1,1),period ='5y')['Close']
```

```python
ew
```

```python
mc = mc.pct_change().dropna()
ew = ew.pct_change().dropna()


mc = (1 + mc['^SP600'] - mc['SPY']).cumprod()
ew = (1 + ew['^R2ESC'] - ew['SPXEW']).cumprod()
```

```python
ew
```

```python
fig = go.Figure()


fig.add_trace(
    go.Scatter(
        x = ew.index,
        y = ew,
        name = 'Equal Weight (Small - Large)'
    )
)

fig.add_trace(
    go.Scatter(
        x = mc.index,
        y = mc,
        name = 'Market Cap Weighted (Small - Large) '
    )
)

fig.update_layout(title = 'Small - Large MC vs EW')
fig.update_yaxes(title = 'Cumulative Returns')
fig.show()
```

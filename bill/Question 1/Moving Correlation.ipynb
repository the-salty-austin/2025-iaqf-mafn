```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
SPY = pd.read_csv("SPY.csv", index_col = 0)
NDXT = pd.read_csv("NDXT.csv", index_col = 0)
```

```python
data = SPY
data['Adj Close_y'] = NDXT['Adj Close']
data.dropna(inplace = True)
```

```python
window = 200
data['rolling_corr'] = data['Adj Close'].rolling(window = window).corr(data['Adj Close_y'])
data.dropna(inplace = True)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['rolling_corr'],
    mode='lines',
    name='Rolling Correlation',
    line=dict(width=2)
))

fig.update_layout(
    title='Rolling Correlation',
    xaxis_title='Date',
    yaxis_title='Rolling Correlation',
    template='plotly_white',
    hovermode='x unified',
    font=dict(size=12)
)

fig.show()
```

```python
data['return_x'] = data['Adj Close'].pct_change()
data['return_y'] = data['Adj Close_y'].pct_change()
data.dropna(inplace = True)
```

```python
window = 200
data['rolling_corr'] = data['return_x'].rolling(window=window).corr(data['return_y'])
data.dropna(inplace = True)
```

```python
data['AAPL'] = AAPL['Return'] * AAPL['MKT_CAP_ARD']
data['AMZN'] = AMZN['Return'] * AMZN['MKT_CAP_ARD']
data['GOOG'] = GOOG['Return'] * GOOG['MKT_CAP_ARD']
data['META'] = META['Return'] * META['MKT_CAP_ARD']
data['MSFT'] = MSFT['Return'] * MSFT['MKT_CAP_ARD']
data['NVDA'] = NVDA['Return'] * NVDA['MKT_CAP_ARD']
data['TSLA'] = TSLA['Return'] * TSLA['MKT_CAP_ARD']
```

```python
data.dropna(inplace = True)
```

```python
data['Summed'] = (data['AAPL'] + data['AMZN'] + data['GOOG'] + data['META'] + 
                 data['MSFT'] + data['NVDA'] + data['TSLA'])
```

```python
window = 150
data['rolling_corr'] = data['Summed'].rolling(window=window).corr(data['SPY Return'])
data.dropna(inplace = True)
```
